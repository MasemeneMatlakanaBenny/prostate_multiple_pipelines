```python
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


dataset=pd.read_csv("prostate.csv")

label="Target"

train,test=train_test_split(dataset,stratify=dataset[label],test_size=0.2,random_state=42)
```

```python
from deepchecks.tabular import Dataset

train_data=Dataset(train,label=label,cat_features=[])
test_data=Dataset(test,label=label,cat_features=[])
```

```python
model=LogisticRegression(solver="liblinear")

model.fit(train.drop("Target",axis=1),train["Target"])
```

```python
import warnings
import numpy as np

np.Inf=np.inf
np.NINF=np.inf

warnings.filterwarnings("ignore",message="Dataframe.applymap")

from deepchecks.tabular.suites import model_evaluation
from deepchecks.tabular.suites import train_test_validation

model_eval_suite=model_evaluation()

model_eval_suite.run(train_data,test_data,model)
```

```python

from deepchecks.tabular.suites import train_test_validation

train_test_suite=train_test_validation()

train_test_suite.run(train_dataset=train_data,test_dataset=test_data)
```
